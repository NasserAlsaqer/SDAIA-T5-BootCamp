```python
from transformers import pipeline

sentiment_analyzer = pipeline("sentiment-analysis")
```

```python
raw_text = [
            "I loved the visualization of the course",
            "I hate this so much"
]
```

```python
sentiment_analyzer(raw_text)
```

### Tokenizer

```python
from transformers import AutoTokenizer

checkpoint = "distilbert/distilbert-base-uncased-finetuned-sst-2-english"

tokenizer = AutoTokenizer.from_pretrained(checkpoint)
```

```python
tokenizer.vocab
```

```python
tokenizer.vocab_size
```

```python
import tensorflow as tf

token_ids = tokenizer(raw_text, padding=True, truncation=True, return_tensors='tf')
token_ids
```

#### Model

```python
from transformers import TFAutoModel

model = TFAutoModel.from_pretrained(checkpoint)
```

```python
output = model(token_ids)
output
```

```python
output.last_hidden_state.shape
```

Bachbone + head

```python
from transformers import TFAutoModelForSequenceClassification

model = TFAutoModelForSequenceClassification.from_pretrained(checkpoint)
model
```

### post preprocessing

```python
output2 = model(token_ids) # show logits **
output2
```

```python
import tensorflow as tf

predictions = tf.math.softmax(output2.logits, axis=-1)
predictions
```

```python
tf.argmax(predictions, axis=-1)
```

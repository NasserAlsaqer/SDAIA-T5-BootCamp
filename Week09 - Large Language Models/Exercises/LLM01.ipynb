```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen2-0.5B")
model = AutoModelForCausalLM.from_pretrained("Qwen/Qwen2-0.5B")
```

```python
sys_prompt = """

أنت أقوى نموذج لتاريخ الإسلام
تستطيع الإجابة بشكل إحترافي عن التاريخ الإسلامي, الرسول والصحابة وكيف كانت حياتهم
أنت عالم ومدرس للتاريخ الإسلامي لطلاب الجامعة في آخر مستوى لهم في الجامعة
تستطيع الإجابة على الاسئلة باللغة العربية واللغة الانجليزية
لديك علم قوي في الجانب التاريخي بشكل عام
اذا كانت هناك معلومات لا تعرفها قل لا أعلم

"""

user_messages = "من هو الرسول؟"

messages = [
    {"role": "system", "content": sys_prompt},
    {"role": "user", "content": user_messages},
]
```

```python
text = tokenizer.apply_chat_template(
    messages,
    tokenize=False,
    add_generation_prompt=True,
)
text
```

```python
model_inputs = tokenizer([text], return_tensors="pt").to(model.device)
model_inputs
```

```python
generated_ids = model.generate(
    **model_inputs,
    max_new_tokens=512,
)
generated_ids
```

```python
generated_ids = [
    output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
]
```

```python
response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
response
```
